## Variables aleatorias y sus tipos

En base a la definición de variable aleatoria que discutimos, se puede hablar de que cada columna de nuestro dataset es una **variable aleatoria**, y que su valor en cada respuesta es una **realización** de dicha variable. Pero, ¿qué tipo tienen esas variables?

### V.A. numéricas

El salario, la edad, los años de experiencia, son variables aleatorias cuyo rango es un conjunto numérico. Podemos clasificarlas en **continuas** o **discretas**, aunque esa distinción se vuelve difusa cuando trabajamos con datos computacionalmente. ¿Por qué? 

* Datos que en teoría son continuos, se miden de manera discreta. Por ejemplo, los *años* de experiencia, la altura de una persona en *centímetros*.
* Datos que en teoría son continuos, se discretizan a fines prácticos. Por ejemplo, la edad, el salario en pesos argentinos.

Para analizar datos continuos se usan frecuentemente los *histogramas*, como en el caso anterior de los sueldos.

**¡Tip!** Antes de graficar, controlar el rango (ya que *seaborn* intentará crear miles de segmentos si el rango es muy grande) y remover los valores nulos.

### Tipos de **variables** vs. Tipos de **datos**

Tenemos que distinguir dos conceptos con el mismo nombre y significado similar, pero que no son iguales:
 - **Tipo de la VA** es el tipo de valores con los que decidimos *intepretar* las realizaciones.
 - **Tipo de datos** es un concepto de programación que indica en qué formato se representa la información. Cuando asignamos a una variable `age` *del programa de Python* una realización de una variable aleatoria conceptual `profile_age`, esa variable `age` también tiene un *tipo de Python*, por ejemplo `int` o `float`.


```python
# Tomamos la primer edad del DF.
age = df.profile_age.iloc[0]
type(age)
```

*¡Importante!* Hay que tener en cuenta también los límites de la capacidad computacional al momento de representar entidades matemáticas.
* Los números reales siempre son "redondeados" a una representación racional.

¿Por qué es importante saberlo? Porque se pueden producir errores de redondeo u obtener resultados aproximados.

```python
print(type(3), type(3.44), type(1/3))  # 1/3 es un número irracional.

import numpy
print(f"Entero mínimo: {numpy.iinfo('int64').min}")
print(f"Entero máximo: {numpy.iinfo('int64').max}")

# Integer Overflow
# numpy.int64(numpy.iinfo('int64').max) + 1
# Integer Underflow
# numpy.int64(numpy.iinfo('int64').min) - 1
```

# Análisis del impacto de los años de experiencia

Ahora que ya sabemos aproximadamente la forma de nuestros datos, podemos pasar a realizar otra pregunta (otra iteración del proceso de análisis): 

**¿Tener más años de experiencia significa que se cobra más?**

Para responder a esta pregunta, analizamos la probabilidad de que un programador tenga un salario mensual mayor que el promedio, cuando tiene una experiencia mayor que 5 años.



```python
# Salario promedio de un programador.
avg_salary = df[salary_col].mean()
print(f'Salario promedio: {avg_salary}')
```

## Medida de probabilidad

En el teórico vimos que si cada uno de nuestros eventos es independiente e idénticamente distribuido, es decir, que $P(\{\omega_i\})=1/k$, entonces la probabilidad de un conjunto $A \subset \Omega$ es la proporción de $A$, donde


$$P(\{\omega_i\})=1/k \implies P(A)=|A|/|\Omega|=|A|/k$$


En este problema en particular, $\Omega$ son todas las respuestas del dataset, cada $a_i$ es una variable que representa una respuesta, y el conjunto $A$ son las respuestas (filas) en la que la columna `salary_col` tiene un valor mayor que el promedio.

```python
filtroSalario = df[salary_col] >= avg_salary
# Probabilidad de tener un salario mayor al promedio.
p_above_avg = len(df[filtroSalario]) / len(df)
print(f'P(salario mayor al promedio) = {p_above_avg}')
```

Por lo tanto, concluimos que:

$$A = \{ salario\_mayor\_al\_promedio \} \implies P(A) = 0.33$$

* ¿Por qué podemos usar la teoría de la probabilidad?

La encuesta se modela como un **experimento aleatorio**.

* ¿Por qué calculamos una probabilidad con esta fórmula?

Se asume que cada *respuesta* es **independiente** e **idénticamente distribuida**.

* ¿Cómo podemos interpretar esta probabilidad?

Es la proporción de respuestas donde ocurre el fenómeno estudiado.

## Probabilidad condicional

Ahora podemos pasar a hablar de la probabilidad condicional entre los dos eventos. La definimos como

$$P(A|B) = \frac{P(A \cap B)}{P(B)}$$

Esto es equivalente a:

$$P(A|B) = \frac{|A \cap B|}{|B|}$$

## Ejercicio

Reponder: **¿Si uno tiene más de 5 años de experiencia, la probabilidad de cobrar más que el promedio aumenta? ¿Estos eventos, son independientes?**

```python
filtroExperiencia = df.profile_years_experience > 5
# Probabilidad de tener más de 5 años de experiencia.
p_years_experience = len(df[filtroExperiencia]) / len(df)
print(f'P(experiencia mayor a 5 años) = {p_years_experience}')
```

Por lo tanto, concluimos que:

$$B = \{ experiencia\_mayor\_cinco \} \implies P(B) = 0.59$$

```python
filtroInterseccion = filtroSalario & filtroExperiencia
# Probabilidad de tener más de 5 años de experiencia, y tener un salario mayor al promedio.
p_intersection = len(df[filtroInterseccion]) / len(df)
print(f'P(intersección) = {p_intersection}')
```

Por lo tanto, concluimos que:

$$P(A \cap B) = 0.27$$

```python
# Probabilidad de tener un salario mayor al promedio, si tiene más de 5 años de experiencia.
p_above_avg_given_experience = p_intersection / p_years_experience
print(f'P(condicional) = {p_above_avg_given_experience}')
```

Por lo tanto, concluimos que:

$$P(A | B) = \frac{P(A \cap B)}{P(B)} = 0.46$$

**Respuesta** Si, si uno tiene más de 5 años de experiencia, la probabilidad de cobrar más que el promedio aumenta.

```python
# ¿Estos eventos son independientes?
print(f'P(intersección) = {p_intersection}')
print(f'P(A) * P(B) = {p_above_avg * p_years_experience}')
```

Por lo tanto, concluimos que:

$$P(A) * P(B) = 0.19$$

**Respuesta** No, estos eventos no son independientes.
